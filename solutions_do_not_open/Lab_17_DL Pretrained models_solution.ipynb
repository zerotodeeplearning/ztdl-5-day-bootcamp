# Pretrained models

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
```

```python
input_path = '../data/generator/class 0/squirrel.jpeg'
```

```python
img = image.load_img(input_path, target_size=(224, 224))
```

```python
img
```

```python
img_array = image.img_to_array(img)
```

```python
img_array.shape
```

```python
plt.imshow(img_array.astype('uint8'))
```

```python
img_tensor = np.expand_dims(img_array, axis=0)
```

```python
img_tensor.shape
```

## Predict class with pre-trained resnet50

```python
from tensorflow.keras.applications.resnet50 import preprocess_input as ppr50
```

```python
img_ppr50 = np.copy(img_tensor)
img_ppr50 = ppr50(img_ppr50)
```

```python
img_ppr50[0].shape
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import decode_predictions as dpr50

model = ResNet50(weights='imagenet')
```

```python
model.summary()
```

```python
preds = model.predict(img_ppr50)
```

```python
preds.shape
```

```python
dpr50(preds, top=3)[0]
```

```python
from tensorflow.keras.applications.vgg16 import preprocess_input as ppv16
```

```python
model.summary()
```

```python
preds = model.predict(img_vgg16)
```

```python
dpv16(preds, top=3)[0]
```
